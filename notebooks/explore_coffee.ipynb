```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
print(coffee.head())
```

```python
# TODO: describe your numeric columns
print(coffee.describe())
```

```python
# TODO: display the null values in your dataframe
print(coffee.isnull().sum())
```

```python
# TODO: drop your null values

coffee_clean = coffee.dropna()
```

```python
# TODO: Select only Minas Gerais

coffee_mg = coffee[coffee['subdivision'] == 'Minas Gerais']
```

```python
# TODO: BEGIN YOUR EDA!

# Check structure
print(coffee_mg.shape)
print(coffee_mg.columns)
print(coffee_mg.dtypes)
```

```python
# Univariate EDA on Coffee Data (Minas Gerais)

# Coffee Production Over Time (60kgs_bag)

# Plot total coffee production in 60kg bags by year
plt.figure(figsize=(10, 5))
sns.lineplot(data=coffee_mg, x='year', y='60kgs_bag', marker='o')

plt.title('Coffee Production in Minas Gerais (60kg Bags)')
plt.xlabel('Year')
plt.ylabel('Bags (60kg)')
plt.tight_layout()
plt.show()
```

```python
# Number of Bearing Trees Over Time

# Plot bearing trees over time
plt.figure(figsize=(10, 5))
sns.lineplot(data=coffee_mg, x='year', y='bearing_trees', marker='o')

plt.title('Number of Bearing Coffee Trees Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Bearing Trees')
plt.tight_layout()
plt.show()
```

```python
#  Number of Non-Bearing Trees Over Time

# Plot non-bearing trees over time
plt.figure(figsize=(10, 5))
sns.lineplot(data=coffee_mg, x='year', y='nonbearing_trees', marker='o')

plt.title('Number of Non-Bearing Coffee Trees Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Non-Bearing Trees')
plt.tight_layout()
plt.show()
 
```

```python
# Bearing Hectares Over Time

# Plot bearing hectares over time
plt.figure(figsize=(10, 5))
sns.lineplot(data=coffee_mg, x='year', y='bearing_hectares', marker='o')

plt.title('Bearing Hectares of Coffee Land Over Time')
plt.xlabel('Year')
plt.ylabel('Bearing Hectares')
plt.tight_layout()
plt.show()
```

```python
# Histogram of bearing trees per hectare
plt.figure(figsize=(7, 4))
sns.histplot(data=coffee_mg, x='bearing_trees_per_hectare', kde=True)

plt.title('Distribution of Bearing Trees Per Hectare')
plt.xlabel('Trees Per Hectare')
plt.tight_layout()
plt.show()
```

```python
# Distribution of bearing_trees, bearing_hectares, and 60kgs_bag

# Plot histograms for all 3 variables
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(data=coffee_mg, x='bearing_trees', kde=True, ax=axes[0])
axes[0].set_title('Distribution of Bearing Trees')

sns.histplot(data=coffee_mg, x='bearing_hectares', kde=True, ax=axes[1])
axes[1].set_title('Distribution of Bearing Hectares')

sns.histplot(data=coffee_mg, x='60kgs_bag', kde=True, ax=axes[2])
axes[2].set_title('Distribution of Coffee Bags (60kg)')

plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis for Coffee Data (Minas Gerais)

# Bearing Trees vs 60kg Bags (Yield Output)

# Scatter plot to see if more bearing trees lead to more production
plt.figure(figsize=(8, 5))
sns.scatterplot(data=coffee_mg, x='bearing_trees', y='60kgs_bag')

plt.title('Coffee Production vs Number of Bearing Trees')
plt.xlabel('Bearing Trees')
plt.ylabel('60kg Coffee Bags')
plt.grid(True)
plt.tight_layout()
plt.show()

# Pearson correlation
corr1 = coffee_mg['bearing_trees'].corr(coffee_mg['60kgs_bag'])
print(f"Pearson correlation between bearing trees and 60kg bags: {corr1:.3f}")
```

```python
# Bearing Trees Per Hectare vs 60kg Bags (Yield Density)

# Explore if tree density affects coffee yield
plt.figure(figsize=(8, 5))
sns.scatterplot(data=coffee_mg, x='bearing_trees_per_hectare', y='60kgs_bag')

plt.title('Coffee Production vs Bearing Trees Per Hectare')
plt.xlabel('Bearing Trees per Hectare')
plt.ylabel('60kg Coffee Bags')
plt.grid(True)
plt.tight_layout()
plt.show()

# Pearson correlation
corr2 = coffee_mg['bearing_trees_per_hectare'].corr(coffee_mg['60kgs_bag'])
print(f"Pearson correlation between bearing trees per hectare and 60kg bags: {corr2:.3f}")
```

```python
# Bearing Hectares vs 60kg Bags

plt.figure(figsize=(8, 5))
sns.scatterplot(data=coffee_mg, x='bearing_hectares', y='60kgs_bag')
plt.title('Bearing Hectares vs 60kg Coffee Bags')
plt.xlabel('Bearing Hectares')
plt.ylabel('Coffee Outpur (60kg Bags)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Pearson correlation
corr3 = coffee_mg['bearing_hectares'].corr(coffee_mg['60kgs_bag'])
print(f"Pearson correlation between bearing hectares and 60kg bags: {corr3:.3f}")
```
